## Question_no_3
### Differentiate between function and method?
## Answer:
A **method** refers to a function which is part of a class. You access it with an instance or object of the class. A **function** doesn’t have this restriction: it just refers to a standalone function. This means that all methods are functions, but not all functions are methods.<br>
Consider this example, where you first define a function plus() and then a addition class with a sum() method:
<code>
#Define a function `plus()`
def plus(a,b):
  return a + b
  
#Create a `addition` class
class addition(object):
  def sum(self, a, b):
    self.contents = a + b
    return self.contents 
</code>
If you now want to call the sum() method that is part of the addition class, you first need to define an instance or object of that class. So, let’s define such an object:
<code>
sumIns = addition()
sumIns.sum(1,2)
</code>

## Question_no_5
### Write a code in python in which create a class named it Car which have 5 attributes such like (model, color and name etc.) and 3 methods. And create 5 object instance from that class.

### Source Code:

```python
class Car():

    def __init__(self,name,brand):
        self.name = name
        self.brand = brand

    def update_price(self,priceinfo):
        self.price = priceinfo

    def update_type(self, shape):
        self.type = shape

    def update_model(self, Modelno):
        self.model = Modelno

    def update_color(self, colorShade):
        self.color = colorShade

    def Show_Details(self):
        return ('\nwe got "{}" by "{}" company \nwhich is registered in "{}" with "{}" in color \nits type is "{}".\nthe price is {}'.format(self.name,self.brand,self.model,self.color,self.type,self.price))

# instance of CAR 1
car1 = Car('Corolla','Toyota')
car1.update_model('2011')
car1.update_type('sun roof')
car1.update_color('Metallic blue')
car1.update_price('PKR 2.9Million')
print(car1.Show_Details())

# instance of CAR 2
car2 = Car('Mustang','Ford')
car2.update_model('2019')
car2.update_type('Reezo Car')
car2.update_color('Shine Red')
car2.update_price('PKR 15.1Million')
print(car2.Show_Details())

# instance of CAR 3
car3 = Car('Civic','Honda')
car3.update_model('2020')
car3.update_type('Type R')
car3.update_color('Fade Grey')
car3.update_price('PKR 4.2 Million')
print(car3.Show_Details())

# instance of CAR 4
car4 = Car('GT-R','Nissan')
car4.update_model('2018')
car4.update_type('Road Show')
car4.update_color('ASH Grey')
car4.update_price('PKR 27.1Million')
print(car4.Show_Details())

# instance of CAR 5
car5 = Car('Enzo','Ferrari')
car5.update_model('2004')
car5.update_type('Road Rage')
car5.update_color('Shine Red')
car5.update_price('PKR 33.1Million')
print(car5.Show_Details())
```
