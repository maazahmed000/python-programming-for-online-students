```python
#Using Dictionary
TodoList = {}
key = 0
print(TodoList)
print('What you wanna do' + "\n" + '1: Add\t2: Del\t3: Delete All\t4:Exit' + '\n')
for repeat in range(50000):
    command = input('Your Command : ')
    if command == 'Add' or command == '1':
        Adding = input('Enter the Todo command: ')
        key = key +1
        TodoList.update({key: Adding})
        print('\n'+str(TodoList)+'\n')
    elif command == 'Del' or command == '2':
        deleting = int(input('enter the Num of the task : '))
        del TodoList[deleting]
        print(TodoList)
    elif command == 'Delete All' or command == '3':
        TodoList.clear()
        key = 0
        print(TodoList)
    elif command == 'Exit' or command == '4':
        print('Quiting the tasks..')
        break
print('\n'+'Final List Left', str(TodoList))
```

```python
#Using List
TodoList = []
print(TodoList)
print('What you wanna do' + "\n" + '1: Add\t2: Del\t3: Delete All\t4:Exit' + '\n')
for repeat in range(50000):
    command = input('Your Command : ')
    if command == 'Add' or command == '1':
        Adding = input('Enter the Todo command: ')
        TodoList.append(Adding)
        print('\n'+str(TodoList)+'\n')
    elif command == 'Del' or command == '2':
        deleting = int(input('Index of the task which has to be deleted from the list starting from 0 : '))
        TodoList.pop(deleting)
        print(TodoList)
    elif command == 'Delete All' or command == '3':
        TodoList.clear()
        print(TodoList)
    elif command == 'Exit' or command == '4':
        print('Quiting the tasks..')
        break
print('\n'+'Final List Left', str(TodoList))
```
